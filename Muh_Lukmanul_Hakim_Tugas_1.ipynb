## Case 2
Pada hari raya idul fitri, Pak Ardi ingin memberikan uang ampao kepada keponakannya. Dia memiliki uang Rp $576,000$ yang akan dibagikan ke $9$ keponakannya. Buatlah program python untuk memudahkan Pak Ardi dalam membagi uang ampao agar sama rata kepada keponakannya !

```
# input by user
uang_ampao = int(input("Masukkan total uang ampao koh Ardi: "))
jumlah_keponakan = int(input("Masukkan total keponakan koh Ardi: "))

uang_pembagian = uang_ampao / jumlah_keponakan

print('Uang yang dibagikan ke masing-masing keponakan koh Ardi adalah: ',uang_pembagian)
```

```
# static variabel
uang_ampao = 576000
jumlah_keponakan = 9

uang_pembagian = uang_ampao / jumlah_keponakan

print('Uang yang dibagikan ke masing-masing keponakan koh Ardi adalah: ',uang_pembagian)
```

# Case 3

Buatlah program yang memudahkan kita untuk menginput harga dan secara otomatis program itu menghitung harga pokok sepeda dengan kriteria pajak sbb. Contoh ketika kita menginputkan harga sepeda 800.000 maka berapa harga final yang harus kita bayar?

        Harga (in Rp)                                             Pajak
        >1000000                                                   15 %
        >500000 and <=1000000                                      10%
        <= 500000                                                  5%

```
def fungsi_harga_ppn (harga_pokok = 0):
  if harga_pokok <= 500000:
    harga_ppn = (harga_pokok * 0.05) + harga_pokok
  elif harga_pokok <= 1000000:
    harga_ppn = (harga_pokok * 0.1) + harga_pokok
  else:
    harga_ppn = (harga_pokok * 0.15) + harga_pokok
  return f"{harga_ppn:,}"
```

```
# input by user
harga_pokok = int(input("Masukkan Harga Pokok Sepeda: "))

print()
print('Total harga setelah ditambah dengan ppn adalah = ', fungsi_harga_ppn(harga_pokok))
```

```
# static varialbe
harga_pokok = 800000

print('Total harga setelah ditambah dengan ppn adalah = ', fungsi_harga_ppn(harga_pokok))
```
